```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.mlab as mb


m1 = 400000
m2 = 300000
g = 6.67384 * 10**-11
r = 3.2
f = g * m1 * m2 / r**2

err_m1 = 500
err_m2 = 1000
err_r = 0.01

err_M1 = 20000
err_M2 = 100000
err_R = 1



def Distf(err_m1, m1, err_m2, m2, err_r, r):
    dist_m1 = np.random.standard_normal(1000000)*err_m1 + m1
    dist_m2 = np.random.standard_normal(1000000)*err_m2 + m2
    dist_r = np.random.standard_normal(1000000)*err_r + r
    return(dist_m1 * dist_m2 * g / dist_r**2)


def Plot(f, err_f, dist_f):
    x = np.arange(f - 5*err_f, f + 5*err_f, 0.00001)
    plt.plot(x, mb.normpdf(x, f, err_f))
    plt.hist(dist_f, normed=True, bins=20)
    plt.show()
```

```python
err_f = f * np.sqrt((err_m1/m1)**2 + (err_m2/m2)**2 + 4 * (err_r/r)**2)
Plot(f, err_f, Distf(err_m1, m1, err_m2, m2, err_r, r))
```

```python
err_f = f * np.sqrt((err_M1/m1)**2 + (err_M2/m2)**2 + 4 * (err_R/r)**2)
Plot(f, err_f, Distf(err_M1, m1, err_M2, m2, err_R, r))
```
